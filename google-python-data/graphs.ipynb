```python
import json
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats

for i in range(10):
    with open(f'best_prompts_similarity{i}.json', 'r') as infile:
        data = json.loads(infile.read())
        queries = np.array(data['queries'])
        lengths = np.array(list(map(len, queries[:, 3])))
        bleu_scores = queries[:, 0].astype(float)
        print(data["prompt"]["prompt"])
        
        # bleu score
        print("bleu_scores")
        rho, val = stats.spearmanr(lengths, bleu_scores)
        print(f'rho: {rho}, val: {val}')
        plt.scatter(lengths, bleu_scores)
        plt.show()
        
        # jaccard
        print("jaccard scores")
        jaccard_scores = queries[:, 1].astype(float)
        rho, val = stats.spearmanr(lengths, jaccard_scores)
        print(f'rho: {rho}, val: {val}')
        plt.scatter(lengths, jaccard_scores)
        plt.show()
        
        # cosine
        print("cosine scores")
        cosine_scores = queries[:, 2].astype(float)
        rho, val = stats.spearmanr(lengths, cosine_scores)
        print(f'rho: {rho}, val: {val}')
        plt.scatter(lengths, cosine_scores)
        plt.show()
```
